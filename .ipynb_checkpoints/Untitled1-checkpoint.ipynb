```python
from keras.models import Sequential
from keras.layers import Dense,Activation,Dropout
from keras.optimizers import Adam,SGD
from keras.regularizers import l2
from keras.metrics import binary_accuracy
import numpy as np
import pandas as pd
from sklearn.metrics import roc_curve,auc,roc_auc_score,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
seed = 11
np.random.seed(seed)

data = pd.read_csv('adult_proc.csv')
data.reindex(np.random.permutation(data.index))
#use input data X to predict income level y
X = data.ix[:,1:87].values
y = data.ix[:,88].values
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.4,random_state=seed)
```

```python
#build and compile the model
model = Sequential()
model.add(Dense(43,init='lecun_uniform',input_dim=86,activation='relu'))
model.add(Dense(12,init='lecun_uniform',activation='sigmoid'))
model.add(Dense(output_dim=1,init='lecun_uniform',activation='sigmoid'))
adm = Adam(lr=.01)
sg = SGD(lr=.001,momentum=0.0,decay=0.0001,nesterov=False)
model.compile(loss='binary_crossentropy',optimizer=sg,metrics=['accuracy'])

#train, reserving some data for validation
model.fit(X_train,y_train,validation_split=.4,nb_epoch=50,batch_size=100,verbose=2,shuffle=True)

y_score = model.predict(X_test)
print ('roc_auc:',roc_auc_score(y_test,y_score))
print ('acc:',accuracy_score(y_test,y_score.astype(int)))
```
